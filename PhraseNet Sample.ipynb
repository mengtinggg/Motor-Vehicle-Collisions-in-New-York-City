```python
import pandas as pd
import numpy as np
import json
from itertools import combinations as c
```

```python
df = pd.read_csv("Motor_Vehicle_Collisions_-_Crashes.csv")
df.head()
```

### Get just vehicles collided, borough and number of injuries/deaths

```python
filtered = df[['BOROUGH']].join(df.loc[:, "VEHICLE TYPE CODE 1":"VEHICLE TYPE CODE 5"]).join(df.loc[:, "NUMBER OF PERSONS INJURED":"NUMBER OF MOTORIST KILLED"])
filtered
```

```python
# drop NA for borough, drop NA for Vehicle Type Code 1 & drop NA for Vehicle Type Code 2
filtered = filtered[filtered['BOROUGH'].notna() & filtered['VEHICLE TYPE CODE 1'].notna() & filtered['VEHICLE TYPE CODE 2'].notna()]
filtered.head()
```

# Find weights (impact of injuries/death)

```python
filtered['weight'] = filtered.loc[:,"NUMBER OF PERSONS INJURED":"NUMBER OF MOTORIST KILLED"].sum(axis=1)
filtered.head()
```

```python
# note that there are records with weight = 0 but since we want to visualise the relationship of collisions, we take them out

filtered.weight.value_counts()
```

```python
# drop weight = 0
filtered = filtered[filtered['weight'] > 0]
filtered
```

```python
# stratified sampling

from sklearn.model_selection import train_test_split

filtered, _ = train_test_split(filtered, test_size=0.8, stratify=filtered[['BOROUGH']], random_state = 42)
filtered
```

# Vehicle Influence

```python
vehicles = filtered.loc[:, "BOROUGH":"VEHICLE TYPE CODE 5"]
vehicles
```

#### Convert header into row entries

```python
vehicles_melt = vehicles.melt('BOROUGH', var_name='type', value_name='vehicle').dropna()
vehicles_melt
```

```python
vehicles_melt.groupby(["BOROUGH", "type", "vehicle"]).size()
```

**Get frequency of each vehicle happening in each Borough and Type**

```python
vehicles_melt['freq'] = 1
vehicles_melt = vehicles_melt.groupby(["BOROUGH", "type", "vehicle"]).freq.count().reset_index()

print(f'max: {vehicles_melt.freq.max()} & min: {vehicles_melt.freq.min()}')

## Note that the disparity between the frequencies are too high
```

**To penalise large values, we will log transform the `freq` column.** <br>
We will use log(x + 1) instead to ensure small transformed values wont be rounded off to 0.

```python
import numpy as np

vehicles_melt['freq_log'] = round((np.log(vehicles_melt.freq) + 1) * 1.5, 3)
vehicles_melt.freq_log.describe()
```

#### ID-ing each vehicle per Borough and Type

```python
vehicles_melt['id'] = vehicles_melt.index + 1
vehicles_melt

# groupby set, or match line by line then create a new col -> combination pair + weight
```

#### Group 'BOROUGH as 'zone'

```python
# Group 'BOROUGH' and 'type' as 'zone'

# zoneDict = {
#     "BRONX": "BRX",
#     "BROOKLYN": "BRK",
#     "QUEENS": "QN",
#     "STATEN ISLAND": "STAT",
#     "MANHATTAN": "MAHT"
# }

# # map above dict to corresponding values under BOROUGH2
# vehicles_melt["BOROUGH2"] = vehicles_melt["BOROUGH"].map(zoneDict)

# # Merge BRONX + VEHICLE TYPE CODE 1 as BRX1
# vehicles_melt["zone"] = vehicles_melt.BOROUGH2 + vehicles_melt.type.str[-1]
# vehicles_melt
```

# Find relationships by identifying the source and target vehicles

```python
vehicles_melt
```

```python
# only get the Borough, Vehicle Type Codes and weight
relationship = filtered.loc[:, "BOROUGH":"VEHICLE TYPE CODE 5"].join(filtered[['weight']])
relationship
```

```python
vehicles_melt.loc[(vehicles_melt['BOROUGH'] == "BROOKLYN") & (vehicles_melt['type'] == "VEHICLE TYPE CODE 1") & (vehicles_melt['vehicle'] == "Sedan")].id
```

```python
vehicles_melt.iloc[226]
```

```python
# vehicles.values.tolist() but without the nan values
# each row in vehicles df is in a list
# outputs a list of lists in the format of [[BOROUGH, Vehicle Type Code 1, Vehicle Type Code 2...]]
per_row = [[y for y in x if pd.notna(y)] for x in vehicles.values.tolist()]
per_row
```

```python
# find all the ids of the vehicles involved in a collision by row
id_map = []
for row in per_row:
    row_ids = []
    for i in range(1, len(row)):
        vtype = "VEHICLE TYPE CODE " + str(i)
        row_ids += list(vehicles_melt.loc[(vehicles_melt["BOROUGH"] == row[0]) & (vehicles_melt['type'] == vtype) & (vehicles_melt['vehicle'] == row[i])].id)
    id_map.append(row_ids)
id_map
```

```python
# link up all involved vehicles by doing combinations
# for example, if ids = [1,2,3], it means vehicle 1,2,3 are involved in the crash
# combi(ids) = [(1,2), (1,3), (2,3)] -> all vehicles can be mapped to each other

id_combis = [list(c(ids, 2)) for ids in id_map]

# import pickle

# with open('id_combis.pkl', 'wb') as f:
#     pickle.dump(id_combis, f)
```

```python
relationship['ids'] = id_map
relationship['ids combined'] = id_combis
relationship
```

```python
# separate lists of tuples into individual rows
relationship = relationship.explode("ids combined")
relationship
```
